```python
def convert3(obj):
    l=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            l.append(i['name'])
            counter+=1
        else:
            break
    
        
    return l
```

### Stemming

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
#porterstemmer is class
#ps is object
ps= PorterStemmer()
```

Helper function

```python
def stem(text):
    y=[]

    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
ps.stem('loved')
```

```python
ps.stem('loving')
```

sorting happended on basis on index, we don't want that

```python
sorted(list(enumerate(similarity[0])), reverse=True, key= lambda x:x[1])
#we are saying, sort on basis of second number
```

Top 5 movies

```python
sorted(list(enumerate(similarity[0])), reverse=True, key= lambda x:x[1])[1:6]
```

steps  
1. find index
2. sort vector with enumarate function
3. select top nearest movies

```python
def recommend(movie):
    movie_index= new_df[new_df['title']==movie].index[0]
    distances= similarity[movie_index]
    movies_list= sorted(list(enumerate(distances)), reverse=True, key= lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
    
    
```

```python
recommend('Avatar')
```

```python
new_df.iloc[1216].title
```

```python
recommend('Batman Begins')
```
